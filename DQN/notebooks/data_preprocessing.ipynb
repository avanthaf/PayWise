# Exploring the datasets

```python
import pandas as pd
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent
RAW = PROJECT_ROOT / "data" / "raw"
```

## DATASET 1: Lending Club
Extracting all the columns

```python
infile = RAW / "lending_club_loan_data.csv"
df = pd.read_csv(infile, low_memory=False)
print(df.columns.tolist())
```

Selected columns from the dataset

```python
cols = ["loan_amnt", "int_rate", "installment"]
```

Checking if the datatype in the selected columns are ideal

```python
print(df[cols].dtypes)
```

Checking the missing values and non negatives

```python
assert df[cols].isna().sum().sum() == 0
assert (df[cols] >= 0).all().all()
```

## DATASET 3: personal_finance_tracker_dataset
Extracting all the columns

```python
infile = RAW / "personal_finance_tracker_dataset.csv"
df = pd.read_csv(infile, low_memory=False)
print(df.columns.tolist())
```

Selected columns from the dataset

```python
cols = ["date",
        "user_id",
        "monthly_income",
        "monthly_expense_total",
        "loan_payment",]
```

Checking if the datatype in the selected columns are ideal

```python
print(df[cols].dtypes)
```

Checking the missing values and non negatives

```python
assert df[["user_id",
           "monthly_income",
           "monthly_expense_total",
           "loan_payment"]].isna().sum().sum() == 0
assert (df[["user_id",
        "monthly_income",
            "monthly_expense_total",
            "loan_payment"]] >= 0).all().all()
```
